**In addition to the bars, we can also add a density estimate**

- **Kernel Density Estimation** (**KDE**) visualizes the distribution of data over a continuous interval. 
- The conventional scale for KDE is: **Total frequency of each bin Ã— Probability**

```python
sns.histplot(data=df, x='price', kde=True);
```

```python
sns.histplot(data=df, x='price', bins=700, kde=True);
```
